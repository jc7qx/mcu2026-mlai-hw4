## 簡介

Pandas是一個Python語言的程式庫，提供資料分析的能力。資料處理經常以表格型式處理資料，如sql的關聯表或Excel的表格。Pandas提供載入、處理、計算、及分析表格化的資料功能，與matplotlib及seaborn結合提供資料視覺化能力。

## Pandas 資料結構
Pandas提出二類型資料結構以便儲存及處理資料，Data Frame及Series。
- Data Frame 以「資料表」形式儲存資料，一筆資料(可稱為entry, record)包括「列」(row)與「行」(column)，代表屬性(attribute)
- Series 代表一個欄位的資料，可視為一維（1-D）資料類型，在Data Frame中的每一欄資料有欄位名稱及列引號為Series，以`df.屬性`或`df['屬性']`表示

## DataFrame

## map, apply, applymap pandas 函式
在 Pandas 中，map、applymap 和 apply 是三種常用於資料轉換的方法。從 Pandas 2.1.0 版本開始，applymap 已被棄用，並由 map 取代其在 DataFrame 上的功能

- Map: 適用於 Series 物件，在較新版本（2.1.0+）中，DataFrame.map 已取代 applymap，成為 DataFrame 執行元素級操作的標準方法。一次處理一個元素。map非常靈活，除了接受函數（callable）外，還能接受 字典（dict） 或另一個 Series 來進行數值映射。
。如果字典中找不到對應的鍵，結果會轉為 NaN
- Apply: 可針對series或dataframe，且可選擇軸向進行。一次處理一個元素。apply 通常以整行或整列為單位進行操作
。傳遞給函數的是一個 Series（該行或該列的所有數據）。apply：通常僅接受 函數（callable）。
- Applymap: 目前已不建議使用，應改用 DataFrame.map

如果 Pandas 有內建的向量化函數（如 str.upper() 或數學運算），應優先使用，因為 apply 系列方法本質上是在進行迴圈運算，效能較慢。 apply 最為全能，可以傳遞額外參數（kwargs），並能決定回傳的是標量、Series 或 DataFrame
。

###Lambda 函數
被視為一種 「可調用對象」（Callable），可以作為參數傳遞給 map、apply 或 applymap 等方法。其標準格式為： lambda 變數: 表達式。根據來源資料，lambda 常用於以下場景：
- 對每個元素轉換：在 map 中對 Series 的每個數值進行運算。例如將數值乘2倍：s.map(lambda x: 2*x)
- 跨欄位運算：在 apply 中處理整行或整列。例如計算每一行的最大值與最小值之差：df.apply(lambda s: s.max() - s.min(), axis=1)
- 新增欄位：在 assign 方法中根據現有數據計算新欄位。例如：df.assign(Area=lambda df: df.Length * df.Height)
- 字典查找與預設值：處理複雜的字典映射邏輯，例如：lambda dic: dic.get('word', 0)

儘管 lambda 很方便，但在處理大型數據時，apply(lambda ...) 的速度通常慢於 Pandas 內建的向量化運算（Vectorized operations），因為它在本質上仍是在進行 Python 層級的迴圈運算

```python
sales_data = pd.DataFrame({
    'Product': ['Apple', 'Banana', 'Cherry'],
    'Price': ["1.2", "0.8", "2.5"]
})
```

```python
sales_data
```

```python
# 2. 定義轉換流程中的各個步驟函數 [2]
def format_price(price):
    return float(price)  # 第一步：將字串轉為浮點數
```

```python
def calculate_tax(price):
    tax_rate = 0.1
    return price * (1 + tax_rate)  # 第二步：計算 10% 稅金
```

```python
def apply_discount(price):
    discount_rate = 0.2
    return price * (1 - discount_rate)  # 第三步：套用 20% 折扣
```

```python
# 3. 呼叫 map 進行多步驟鏈式轉換 [2]
# map 會由左至右依序執行轉換
sales_data['Formatted_Price'] = (
    sales_data['Price']
    .map(format_price)
    .map(calculate_tax)
    .map(apply_discount)
)
```

```python
print(sales_data)
```

##Groupby, Merge, Pivot Table

```python
import pandas as pd
import numpy as np

#建立展示資料集
data = {
    'date': pd.date_range('2024-01-01', periods=100, freq='D'),
    'product': np.random.choice(['A', 'B', 'C'], 100),
    'region': np.random.choice(['North', 'South', 'East', 'West'], 100),
    'sales': np.random.randint(100, 1000, 100),
    'costs': np.random.randint(50, 500, 100)
}
df = pd.DataFrame(data)
df
```

groupby() 方法可以根據一個或多個欄位對資料進行分組，並對每個群組應用聚合或其他操作。

Agg函數是GroupBy操作的一個強大擴展，它允許我們在一次操作中對多個列應用多個聚合函數。 Agg函數的靈活性使得它成為數據分析中不可或缺的工具。

```python
summary = df.groupby(['product', 'region']).agg({
    'sales': ['sum', 'mean', 'count'],
    'costs': ['sum', 'mean'],
    'date': ['min', 'max']
})
summary
```

```python
products = pd.DataFrame({
    'product': ['A', 'B', 'C'],
    'category': ['Electronics', 'Furniture', 'Clothing'],
    'margin': [0.3, 0.25, 0.4]
})
products
```

merge 方法用於合併兩個 DataFrame 的方法，類似於 SQL 的 JOIN 操作。它可以根據指定的鍵（key）進行內合併、左合併、右合併或外合併

- 內合併 (Inner Join)：僅保留兩個 DataFrame 中鍵值匹配的資料。
- 左合併 (Left Join)：保留左側 DataFrame 的所有資料，右側無匹配則填入 NaN。
- 右合併 (Right Join)：保留右側 DataFrame 的所有資料，左側無匹配則填入 NaN。
- 外合併 (Outer Join)：保留所有資料，無匹配的部分填入 NaN

```python
#合併兩個 DataFrame(df, products)，以 product 欄位作為鍵值（key）來對應，
#左連接 (left join)，意思是保留 df 裡的所有資料，並把 products 裡對應的欄位加進來。
#如果 df 裡有某些產品在 products 沒有對應，這些新欄位會是 NaN
enriched = df.merge(products, on='product', how='left')
enriched[['product', 'sales', 'category', 'margin']].head(10)
```

Pandas 的 pivot_table 方法是一個強大的工具，用於創建類似 Excel 樞紐分析表的數據透視表，幫助進行數據分組、匯總和統計分析。

https://www.learncodewithmike.com/2022/02/pandas-pivot-table.html

```python
#sales是要彙整的數值欄位，product 欄位作為列索引（每一列代表一個產品）。
#region 欄位展開成欄位（每一欄代表一個地區）。
#當同一產品在同一地區有多筆銷售紀錄時，會用「加總」來彙整。
#如果某產品在某地區沒有銷售紀錄，填入 0 而不是 NaN。
pivot = df.pivot_table(
    values='sales',
    index='product',
    columns='region',
    aggfunc='sum',
    fill_value=0
)
pivot
```

```python
df['profit'] = df['sales'] - df['costs']
#df新增一個margin_pct欄位，apply會對DataFrame的每一列（row）套用一個函數(axis=1)
#匿名函數，輸入是一列資料(row)
df['margin_pct'] = df.apply(
    lambda row: (row['profit'] / row['sales'] * 100) if row['sales'] > 0 else 0,
    axis=1
)
df[['sales', 'costs', 'profit', 'margin_pct']]
```

```python
#向量化運算
import numpy as np

df['margin_pct'] = np.where(
    df['sales'] > 0,
    df['profit'] / df['sales'] * 100,
    0
)
df[['sales', 'costs', 'profit', 'margin_pct']]
```
